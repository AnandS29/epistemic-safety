```python
import numpy as np
import matplotlib.pyplot as plt
import nashpy as nash
import itertools
```

```python
# Defining MDP

states = ["s1", "s2", "s3", "s4"]

human_actions = [("c",0), ("c",1), ("d",0), ("d",1)]
robot_actions = [("c",0), ("c",1), ("d",0), ("d",1)]

def transition(s, uH, uR):
    if s == "s1":
        if 0 in uR:
            return "s2"
        elif 1 in uR:
            return "s1"
    elif s == "s2":
        if 0 in uH:
            return "s3"
        elif 1 in uH:
            return "s1"
    elif s == "s3":
        return "s3"
    elif s == "s4":
        if 1 in uH and 1 in uR:
            return "s2"
        else:
            return "s4"
    else:
        print("Incorrect state")
        
def prisoners_dilemma(uH,uR):
    if uH == "c" and uR == "c":
        return (3,3)
    if uH == "d" and uR == "d":
        return (1,1)
    if uH == "c" and uR == "d":
        return (0,5)
    if uH == "d" and uR == "c":
        return (5,0)
    
def pay_or_punish(uH,uR):
    if uR == "c":
        return (10,10)
    if uR == "d":
        return (-10,-10)
    
def social(t):
        return (t[0],t[1],t[0]+t[1])
        
def reward(s, uH, uR):
    if s == "s1":
        r = prisoners_dilemma(uH[0], uR[0])
    elif s == "s2":
        r = prisoners_dilemma(uH[0], uR[0])
    elif s == "s3":
        r = pay_or_punish(uH[0],uR[0])
    elif s == "s4":
        r = prisoners_dilemma(uH[0], uR[0])
    else:
        print("Incorrect state")
        return None
    return social(r)
```

```python
T = 5
gamma = 0.99
v_funs = {i:{s:None for s in states} for i in range(T+1)}
v_funs_adv = {i:{s:None for s in states} for i in range(T+1)}
v_funs_adv_p = {i:{s:None for s in states} for i in range(T+1)}
uHs = {i:{s:{a:None for a in human_actions} for s in states} for i in range(T)}
```

```python
# Initialize value functions
for s in states:
    v_funs[T][s] = 0
    v_funs_adv[T][s] = 0
    v_funs_adv_p[T][s] = 0
```

```python
# Compute baseline adversarial value function
for t in range(T):
    for s in states:
        v_funs_adv[t][s] = 0
```

```python
# Compute value function
t = T-1
while t >= 0:
    for s in states:
        
        # Find adversary to policy for human's reward
        
        
        # Find cooperative action subject to constraints
        qS = None
        uH_opt = None
        for uH in human_actions:
            for uR in robot_actions:
                # Check if action passes constraints
                if False:
                    continue
                r = reward(s, uH, uR)
                qS_temp = r[0] + gamma*v_funs[t+1][transition(s,uH,uR)] # social value of action
                if qS is None or qS_temp > qH:
                    qS = qS_temp
                    uH_opt = uH
        
        v_funs[t][s] = qS
        uHs[t][s][uH_opt] = 1.0
    t = t - 1
```
